# Santander 은행 고객 만족 예측

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split

df = pd.read_csv('santander-customer-satisfaction/santander_train.csv')
df.head()
```

## 만족, 불만족 비율 확인해보기

```python
print(df['TARGET'].value_counts())
unsatisfied = df[df['TARGET']==1]['TARGET'].count()
satisfied = df[df['TARGET']==0]['TARGET'].count()
print(f'불만족 비율은 {unsatisfied/satisfied : .2f}')
```

### 열이 워낙에 많기 때문에 일일이 결측치를 확인할 수 없지만 var3의 경우 min 값에 이상치가 있음을 한눈에 파악할 수 있음

```python
df.describe()
```

### var3의 값 분포 확인

```python
df['var3'].value_counts()[:10]
```

### 이상치를 가장 많은 값인 2로 바꿔줌
> 데이터프레임에도 replace함수를 이용하여 값을 일괄적으로 수정할 수 있음

```python
df['var3'].replace(-999999, 2, inplace=True)
df['var3'].value_counts()[:10]
```

### 0(만족)이 1(불만족)에 비해서 월등히 많기 때문에 불균형 데이터임
### train test 셋을 나눈 후 0과 1의 비율이 비슷하게 들어갔는지 확인 필요
> 만약 똑같이하고싶다면 train_test_split에 stratify = y_target 을 넣어주면 됨

```python
X_data = df.iloc[:,:-1]
y_target = df.iloc[:,-1]
X_train, X_test, y_train, y_test = train_test_split(X_data, y_target,
                                                   test_size=0.2,
                                                   random_state=0,
                                                   stratify = y_target)

train_cnt = y_train.count()
test_cnt = y_test.count()

print('학습 셋 target값 비율')
print(y_train.value_counts()/train_cnt)
print('\n 테스트 셋 target값 비율')
print(y_test.value_counts()/test_cnt)
```

### 1. 원래는 evals에 검증 데이터셋을 넣어야하지만 불균형데이터이므로 그냥 test셋을 넣음 
### 2. fit을 한 후 하이퍼 파라미터 수정
> `early_stopping_rounds` : 조기중단 횟수 설정
> `eval_metric` : 성능 평가 지표
> `eval_set` : 검증 데이터로 사용할 튜플형식의 데이터
### 3. ROC AUC score 확인

### (참고)
####  predict_proba를 하게되면 0에 대한 예측확률값, 1에 대한 예측확률값으로 (, 2) 의 shape으로 출력됨
#### fit을 할때 조기중단을 하려면 eval_set, eval_metric이 있어야함
#### 반복마다 eval_set으로 지정된 데이터에서 eval_metric의 지정된 평가 지표로 예측 오류 측정

```python
from lightgbm import LGBMClassifier
from sklearn.metrics import roc_auc_score

lgbm_clf = LGBMClassifier(n_estimators=500)

evals = [(X_test, y_test)]
lgbm_clf.fit(X_train, y_train, early_stopping_rounds=100, 
             eval_metric='auc', eval_set=evals, verbose=True)

lgbm_roc_score = roc_auc_score(y_test, lgbm_clf.predict_proba(X_test)[:,1], 
                              average='macro')
print(f'ROC AUC : {lgbm_roc_score : .2f}')
```
